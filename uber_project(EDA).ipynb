```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data=pd.read_csv('final_internship_data.csv')
```

```python

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='passenger_count')
plt.title('Distribution of Passenger Counts', fontsize=15)
plt.xlabel('Number of Passengers', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='hour')
plt.title('Distribution of hours', fontsize=15)
plt.xlabel('HOUR', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='day')
plt.title('Distribution of days', fontsize=15)
plt.xlabel('DAY', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='weekday')
plt.title('Distribution of weekdays', fontsize=15)
plt.xlabel('WEEKDAY', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='month')
plt.title('Distribution of month', fontsize=15)
plt.xlabel('MONTH', fontsize=12)
plt.ylabel('Count', fontsize=12)


plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='year')
plt.title('Distribution of years', fontsize=15)
plt.xlabel('YEAR', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Weather')
plt.title('Distribution of weather status', fontsize=15)
plt.xlabel('weather', fontsize=12)
plt.ylabel('Count', fontsize=12)


plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Car Condition')
plt.title('Distribution of car condition', fontsize=15)
plt.xlabel('car condition', fontsize=12)
plt.ylabel('Count', fontsize=12)


plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data['fare_amount'], kde=True, bins=30, color='red')
plt.title('Distribution of Fare Amounts')
plt.xlabel('Fare Amount ($)')
plt.ylabel('Number of Trips')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Weather', y='Traffic Condition', alpha=0.5)
plt.title('Relationship: Weather vs Traffic Condition')
plt.xlabel('Weather')
plt.ylabel('Traffic Condition')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Weather', y='distance', alpha=0.5)
plt.title('Relationship: Weather vs distance')
plt.xlabel('Weather')
plt.ylabel('distance')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Traffic Condition', y='distance', alpha=0.5)
plt.title('Relationship: Traffic Condition vs distance')
plt.xlabel('Traffic Condition')
plt.ylabel('distance')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='distance', y='fare_amount', alpha=0.5)
plt.title('Relationship: distance vs fare_amount')
plt.xlabel('distance')
plt.ylabel('fare_amount')
plt.show()
```
